```python
mv.experience = mv.experience.fillna("zero")
mv
```

```python
import math
test_score = math.floor(mv['test_score(out of 10)'].mean())
test_score
```

```python
mv['test_score(out of 10)'] = mv['test_score(out of 10)'].fillna(test_score)
mv
```

```python
reg = linear_model.LinearRegression()
reg.fit(mv[['experience','test_score(out of 10)','interview_score(out of 10)']], mv['salary($)'])
```
